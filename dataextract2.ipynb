```python
import pandas as pd
```

```python
date_list = pd.date_range(start='1/1/2018', end = '1/1/2019', freq='d')
df = pd.DataFrame(dict(date=date_list))
print(df)
```

```python
#winter 1st June to 31 August
#Autum 1st March to 31 May
#Summer 1st December to 28 Feb
#Spring 1st September to 30th November
```

```python
df['Week Number'] = df['date'].dt.week
```

```python
df['Finacial Year'] = df['date'].dt.year
```

```python
df['Date'] = df['date'].dt.day
```

```python
df['month'] = df['date'].dt.month
```

```python
df['Day'] = df['date'].dt.weekday_name
```

```python
df['Month Name'] = df['date'].dt.month_name()
```

```python
#df['DateID'] = df['date'].dt.month_name()
```

```python
#winter 1st June to 31 August
#Autum 1st March to 31 May
#Summer 1st December to 28 Feb
#Spring 1st September to 30th November

df.loc[(df['Month Name'] == "December") |(df['Month Name'] == "January") | (df['Month Name'] == "February"),'season']='Summer'
df.loc[(df['Month Name'] == "March") |(df['Month Name'] == "April") | (df['Month Name'] == "May"),'season']='Autum'
df.loc[(df['Month Name'] == "June") |(df['Month Name'] == "July") | (df['Month Name'] == "August"),'season']='Winter'
df.loc[(df['Month Name'] == "September") |(df['Month Name'] == "October") | (df['Month Name'] == "November"),'season']='Spring'
```

```python
#weekend flag
df.loc[(df['Day'] == "Saturday") |(df['Day'] == "Sunday"),'Weekend Flag']= True
```

```python
#AESDT Falg
df.loc[(df['Month Name']=="October") & (df['Day']=="Sunday")].min()['Date']

#find index
index_list = (df.loc[(df['Month Name']=="October") & (df['Day']=="Sunday")].min()['Date']).index.tolist()
index_list
```
